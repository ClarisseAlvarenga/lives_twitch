```
#criando o boxplot das notas de matemática por renda
def cria_boxplot_por_renda(coluna, titulo, dados):
  plt.figure(figsize=(12,6))
  plt.title(titulo, fontsize=20)
  ax= sns.boxplot(y='Q006', x=coluna, data=dados,order=renda_ordenada )
  ax.set_yticklabels(renda_familiar.values())
  ax.set_ylabel('Renda Familiar')
  plt.show()
```

```
#calculando os limites do intervalo de confiança: (média +- 2x o desvio padrão 95% de confiança)
limite_inferior = media - 2*desvio_padrao
limite_superior = media + 2*desvio_padrao
(limite_inferior, limite_superior)
```

```
print(f'O intervalo de confiança vai de {limite_inferior.round(2)} até {limite_superior.round(2)}')
```

```
#criando uma funcão para calcular o intervalo de confiança do nosso modelo cv
def calcula_mse(resultados):
  media = (resultados['test_score']*-1).mean()
  desvio_padrao = (resultados['test_score']*-1).std()
  limite_inferior = media - 2*desvio_padrao
  limite_superior = media + 2*desvio_padrao
  print(f'O intervalo de confiança vai de {limite_inferior.round(2)} até {limite_superior.round(2)}')

```

```
#chamando nossa função
calcula_mse(cv)
```
